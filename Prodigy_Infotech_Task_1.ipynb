```python
# Drop unnecessary columns
df.drop(['Series Name','Country Code'],axis=1,inplace=True)
```

During the cleaning and data preprocessing phase, we discovered that the dataset contains information on 217 countries. The data has no null values and no duplicates.

The dataset is structured into categories as follows:

The first 217 rows represent the total population of each country.
The next 217 rows represent the female population of each country.
The following 217 rows represent the female population as a percentage population of each country as a total.
The next 217 rows represent the male of the total population.
The final 217 rows represent the male population as a percentage of the total population.


At the next stage, we will sort the data for each 10-year period

```python
total_population_data = df[df['Series Code'] == 'SP.POP.TOTL']

# Sort data based on the total population for 2022
total_population_sorted = total_population_data.sort_values(by="2022", ascending=False)
total_top_ten_countries = total_population_sorted.head(10)
total_less_countries = total_population_sorted.tail(10)

# Sort data based on the total population for 2011
total_population_sorted_2011 = total_population_data.sort_values(by="2011", ascending=False)
total_top_ten_countries_2011 =total_population_sorted_2011.head(10)
total_less_countries_2011 = total_population_sorted_2011.tail(10)

# Sort data based on the total population for 2001
total_population_sorted_2001 = total_population_data.sort_values(by="2001", ascending=False)
total_top_ten_countries_2001 =total_population_sorted_2001.head(10)
total_less_countries_2001 = total_population_sorted_2001.tail(10)
```

```python
# top countries
plt.figure(figsize=(15, 10))

plt.subplot(3, 1, 1)
sns.barplot(data=total_top_ten_countries,x='2022',y='Country Name',palette='viridis')

plt.subplot(3,1,2)
sns.barplot(data=total_top_ten_countries_2011,x='2011',y='Country Name',palette='plasma')

plt.subplot(3,1,3)
sns.barplot(data=total_top_ten_countries_2001,x='2001',y='Country Name',palette='inferno')
```

Over the last 30 years:  

- India's population has increased significantly.  
- China's population growth has been declining.   
- The population growth in the United States and Indonesia has remained relatively stable.  
- Mexico and Nigeria have experienced a significant population increase over the past two decades.  
- Russia Federation's population growth has been decreasing.  
- Japan's population has declined significantly.

```python
# the least countries
plt.figure(figsize=(15, 10))

plt.subplot(3, 1, 1)
sns.barplot(data=total_less_countries,x='2022',y='Country Name',palette='viridis')

plt.subplot(3,1,2)
sns.barplot(data=total_less_countries_2011,x='2011',y='Country Name',palette='plasma')

plt.subplot(3,1,3)
sns.barplot(data=total_less_countries_2001,x='2001',y='Country Name',palette='inferno')
```

Over the last 30 years:  

- The Marshall Islands had the highest population among the smallest countries in 2022.  
- Liechtenstein had the highest population in 2011, followed by Monaco and Sint Maarten (Dutch part).  
- In 2001, Monaco had the largest population, with Sint Maarten (Dutch part) and St. Martin (French part) following.  
- Some countries, like Monaco and Liechtenstein, maintained relatively stable populations.  
- Others, like the Marshall Islands, saw noticeable population growth.  
- The population rankings among these small nations have shifted over time.

```python
total_top_ten_countries['Country Name'] = total_top_ten_countries['Country Name'].str.strip()

# number of males in top 10 countries has population in 2022
df_filtered_male = df[df['Series Code'] == 'SP.POP.TOTL.MA.IN']
df_filtered_male['Country Name'] = df_filtered_male['Country Name'].str.strip()
male_of_top_ten_countries = df_filtered_male[df_filtered_male['Country Name'].isin(total_top_ten_countries['Country Name'])]

# number of females in top 10 countries has population in 2022
df_filtered_female = df[df['Series Code'] == 'SP.POP.TOTL.FE.IN']
df_filtered_female['Country Name'] = df_filtered_female['Country Name'].str.strip()
female_of_top_ten_countries = df_filtered_female[df_filtered_female['Country Name'].isin(total_top_ten_countries['Country Name'])]

total_less_countries['Country Name']=total_less_countries['Country Name'].str.strip()
# number of males in less 10 countries has population in 2022

male_of_the_least_ten_countries =df_filtered_male[df_filtered_male['Country Name'].isin(total_less_countries['Country Name'])]

# number of females in less 10 countries has population in 2022
female_of_the_least_ten_countries=df_filtered_female[df_filtered_female['Country Name'].isin(total_less_countries['Country Name'])]
```

```python
plt.figure(figsize=(16,15))
plt.subplot(2,1,1)
female_population = female_of_top_ten_countries["2022"].values
sns.barplot(x="Country Name", y="2022", data=female_of_top_ten_countries, color="Pink", label="Female Population")
sns.barplot(x="Country Name", y="2022", data=male_of_top_ten_countries, bottom=female_population, color="DodgerBlue", label="Male Population")

plt.subplot(2,1,2)
plt.xticks(rotation=45)
female_population_2 = female_of_the_least_ten_countries["2022"].values
sns.barplot(x="Country Name", y="2022", data=female_of_the_least_ten_countries, color="Pink", label="Female Population")
sns.barplot(x="Country Name", y="2022", data=male_of_the_least_ten_countries, bottom=female_population_2, color="DodgerBlue", label="Male Population")
```
